# <u>DATA PREPROCESSING</u>

Let us clean and encode the data which can be later used for the models.<br>
<br>
<br>
->  Let us remove the columns which are either identical through out the dataset(eg: Over18). Since the values are always same, it won't affect the prediction as much.<br>
->  Let us also remove the columns which are varying throughout the dataset. Since the column contains so many unique values, it will not affect the target column which is Attrition.

```python
#Remove some non predictive columns

emp_attr = emp_attr.drop('Over18', axis = 1)
emp_attr = emp_attr.drop('EmployeeNumber', axis = 1)
emp_attr = emp_attr.drop('StandardHours', axis = 1)
```

```python
emp_attr = emp_attr.drop('EmployeeCount', axis = 1)
```

```python
#Get the correlation matrix

emp_attr.corr()
```

```python
#Vizualing the correlation 

plt.figure(figsize = (14,14))
sns.heatmap(emp_attr.corr(), annot = True, fmt = '.0%', cbar = False)
```

```python
#Converting categorical data into numerical data using Label Encoding

from sklearn.preprocessing import LabelEncoder

for column in emp_attr.columns:
    if emp_attr[column].dtype == np.number:
        continue
    emp_attr[column] = LabelEncoder().fit_transform(emp_attr[column])    
```

```python
#Create a new column

emp_attr['Age_Years'] = emp_attr['Age']
```

### We drop the age column so as to have the attrition column as the first one in the dataset. No deleting or inserting of data is done here.

```python
#Drop age column

emp_attr = emp_attr.drop('Age', axis = 1)
```

#  <u>PREDICTION</u>
We split the data into training and testing data.<br> We then use the training data to train the model we are using, and then use the testing data to implement the model and check its accuracy in prediction.



```python
#Split the data

X = emp_attr.iloc[:, 1:emp_attr.shape[1]].values #We input all the columns except Attrition.
Y = emp_attr.iloc[:, 0].values  #We input our target column Attrition.
```

```python
#Split the data into 75% training and 25% testing data

from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.25, random_state = 0)
```

## (1) Random Forest classifier 

Scikit-Learn implements a set of sensible default hyperparameters for all models, but these are not guaranteed to be optimal for a problem. The best hyperparameters are usually impossible to determine ahead of time, and tuning a model is where machine learning turns from a science into trial-and-error based engineering.

```python
from sklearn.ensemble import RandomForestClassifier
forest = RandomForestClassifier(n_estimators = 10, criterion = 'entropy', random_state = 0)
print('Parameters currently in use:\n')
print(forest.get_params())
```

```python

# Number of trees in random forest
n_estimators = [int(x) for x in np.linspace(start = 200, stop = 2000, num = 10)]
# Number of features to consider at every split
max_features = ['auto', 'sqrt']
# Maximum number of levels in tree
max_depth = [int(x) for x in np.linspace(10, 110, num = 11)]
max_depth.append(None)
# Minimum number of samples required to split a node
min_samples_split = [2, 5, 10]
# Minimum number of samples required at each leaf node
min_samples_leaf = [1, 2, 4]
# Method of selecting samples for training each tree
bootstrap = [True, False]
# Create the random grid
random_grid = {'n_estimators': n_estimators,
               'max_features': max_features,
               'max_depth': max_depth,
               'min_samples_split': min_samples_split,
               'min_samples_leaf': min_samples_leaf,
               'bootstrap': bootstrap}
print(random_grid)
```

```python
# Use the random grid to search for best hyperparameters
# First create the base model to tune
from sklearn.model_selection import RandomizedSearchCV
# Random search of parameters, using 3 fold cross validation, 
# search across 100 different combinations, and use all available cores
forest_best = RandomizedSearchCV(estimator = forest, param_distributions = random_grid, n_iter = 100, cv = 3, verbose=2, random_state=42, n_jobs = -1)
# Fit the random search model
```

```python
#Using the Random Forest Classifier 

forest_best.fit(X_train, Y_train)
Y_test_pred = forest_best.predict(X_test)
```

```python
#Get the accuracy on the training data set

forest_best.score(X_train, Y_train)
```

### The above score is the best score we can obtain since we tuned our hyperparamters to optimize our Random Forest Classifier model.

```python
#Show the confusion matrix and accuracy score for the model on the test data

from sklearn.metrics import confusion_matrix, classification_report

cm = confusion_matrix(Y_test, Y_test_pred)

TN = cm[0][0]
TP = cm[1][1]
FN = cm[1][0]
FP = cm[0][1]

print(cm)
print('Model Testing Accuracy = {}'.format ( (TP + TN) / (TP + TN + FP + FN)))
print ("\nClassification Report on Test Data:\n",classification_report(Y_test, Y_test_pred))
```

### We get an accuracy of 86.6 % using Random Forest Classifier.

## (2) Naive Bayes Classifier 

```python
#Using the Naive Bayes Classifier

from sklearn.naive_bayes import GaussianNB
model = GaussianNB()
model.fit(X_train, Y_train)
Y_test_pred = model.predict(X_test)
```
